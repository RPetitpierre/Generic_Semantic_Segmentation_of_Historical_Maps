# Import

```python
import cv2
from PIL import Image
import numpy as np
from skimage.feature import local_binary_pattern
```

# Script

```python
patch_path = 'dataset/paris/images/12148_btv1b530855043f1_0.png'
image = cv2.cvtColor(cv2.imread(patch_path), cv2.COLOR_BGR2RGB)
Image.fromarray(image)
```

```python
gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
Image.fromarray(gray)
```

```python
_, otsu = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY+cv2.THRESH_OTSU)
Image.fromarray(otsu)
```

```python
lbp_radius = 3
lbp = local_binary_pattern(otsu, 8*lbp_radius, lbp_radius, 'uniform').astype('uint16')
lbp = (lbp*255/np.max(lbp)).astype('uint8')
_, lbp = cv2.threshold(lbp, 0, 255, cv2.THRESH_BINARY+cv2.THRESH_OTSU)
Image.fromarray(lbp)
```
